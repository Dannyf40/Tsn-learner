```python
import os

# Create folder
os.makedirs('/content/home_organiser', exist_ok=True)
os.chdir('/content/home_organiser')

# Create main.py
with open('main.py', 'w') as f:
    f.write('''from kivy.app import App
from kivy.uix.boxlayout import BoxLayout
from kivy.uix.checkbox import CheckBox
from kivy.uix.label import Label
from kivy.uix.button import Button
from kivy.uix.tabbedpanel import TabbedPanel, TabbedPanelItem
from kivy.clock import Clock
import requests
from plyer import notification
import datetime

class RoutineLayout(BoxLayout):
    def __init__(self, **kwargs):
        super().__init__(**kwargs)
        self.orientation = 'vertical'
        tabs = TabbedPanel()
        morning_tab = TabbedPanelItem(text='Morning')
        evening_tab = TabbedPanelItem(text='Evening')

        # Morning
        morning_layout = BoxLayout(orientation='vertical', padding=10, spacing=10)
        for task in ['Bag ready?', 'Sandwiches packed?', 'Books in bag?', 'Uniform on?']:
            row = BoxLayout()
            row.add_widget(CheckBox())
            row.add_widget(Label(text=task, halign='left', size_hint_x=0.8))
            morning_layout.add_widget(row)
        self.weather_label = Label(text='Tap to load weather...', size_hint_y=None, height=50)
        morning_layout.add_widget(self.weather_label)
        weather_btn = Button(text="Get Today\\'s Weather", size_hint_y=None, height=50)
        weather_btn.bind(on_press=self.fetch_weather)
        morning_layout.add_widget(weather_btn)
        self.bus_btn = Button(text='Check Bus Time', size_hint_y=None, height=50)
        self.bus_btn.bind(on_press=self.check_bus)
        morning_layout.add_widget(self.bus_btn)
        morning_tab.add_widget(morning_layout)

        # Evening
        evening_layout = BoxLayout(orientation='vertical', padding=10, spacing=10)
        for task in ['Uniform hanging up?', 'Homework done?', 'Bag prepped?']:
            row = BoxLayout()
            row.add_widget(CheckBox())
            row.add_widget(Label(text=task, halign='left', size_hint_x=0.8))
            evening_layout.add_widget(row)
        evening_tab.add_widget(evening_layout)

        tabs.add_widget(morning_tab)
        tabs.add_widget(evening_tab)
        self.add_widget(tabs)
        self.update_bus_visibility()
        Clock.schedule_interval(self.send_morning_prompt, 60)

    def update_bus_visibility(self):
        today = datetime.date.today()
        if today >= datetime.date(2025, 9, 1):
            self.bus_btn.disabled = False
            self.bus_btn.text = "Check Bus Time (Active)"
        else:
            self.bus_btn.disabled = True
            self.bus_btn.text = "Bus starts in Sept 2025"

    def fetch_weather(self, instance):
        API_KEY = "YOUR_KEY_HERE"
        CITY = "London,UK"
        if API_KEY == "YOUR_KEY_HERE":
            self.weather_label.text = "Add API key!"
            return
        url = f"http://api.openweathermap.org/data/2.5/weather?q={CITY}&appid={API_KEY}&units=metric"
        try:
            data = requests.get(url).json()
            temp = data['main']['temp']
            desc = data['weather'][0]['description'].capitalize()
            self.weather_label.text = f"{temp}°C | {desc}"
        except:
            self.weather_label.text = "Weather: Offline"

    def check_bus(self, instance):
        notification.notify(title="School Bus", message="Next bus at 8:15 AM!")

    def send_morning_prompt(self, dt):
        notification.notify(title="Morning Routine!", message="Time to get ready!")

class HomeOrganiserApp(App):
    def build(self):
        return RoutineLayout()

if __name__ == '__main__':
    HomeOrganiserApp().run()
''')

# Create myapp.kv
with open('myapp.kv', 'w') as f:
    f.write(''':kivy 2.1.0

<RoutineLayout>:
    canvas.before:
        Color:
            rgba: 0.95, 0.95, 0.98, 1
        Rectangle:
            pos: self.pos
            size: self.size

<Label>:
    color: 0.2, 0.2, 0.2, 1
    font_size: '16sp'

<Button>:
    background_normal: ''
    background_color: 0.3, 0.6, 1, 1
    color: 1, 1, 1, 1
    font_size: '16sp'
    size_hint_y: None
    height: 50

<CheckBox>:
    size_hint_x: 0.2
''')

print("Files created: main.py, myapp.kv")
```
